## ğŸ”¹ What is Scope?

**Scope** refers to where a variable is accessible.

| **Built-in (B)**  | Python built-ins                     | `len()`, `print()`  |

## ğŸ”¸ 1. **Local Scope** â€“ Inside a Function

````python
def convert_numbers_flex(data):
    result = []
    for item in data:
        try:
            num = float(item)
            result.append(int(num) if num.is_integer() else num)
            
        except (ValueError, TypeError):
            result.append(item)          # Keep strings or other types
    return result

data = ["12", "abc", "42.5", "100", "xyz", "0.0", "123.00"]
print(convert_numbers_flex(data))
````

````python
#âœ… Topic 2 â€“ Q10: Custom Expression Evaluator Function
#ğŸ¯ Objective:
#Build a Python function that evaluates simple 2-operand expressions using basic arithmetic and comparison operators. This simulates what Pythonâ€™s interpreter does internally â€” and what custom test engines often need.


def evaluate(op, a, b):
    # op is a string like "+", ">", "==", etc.
    # a and b are operands
    scope = ('+','-','/','//','>','<','==')

    if not op in scope:
        print('invalid')
    elif op=='+':
        return float(a)+float(b)
    elif op=='-':
        return float(a)-float(b)
    elif op=='/':
        return float(a)/float(b)
    elif op=='//':
        return float(a)//float(b)
    elif op=='>' and float(a)>float(b):
        return True
    
def evaluate(op, a, b):
    # Convert operands to float for flexibility
    a = float(a)
    b = float(b)

    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        return a / b
    elif op == "//":
        return a // b
    elif op == ">":
        return a > b
    elif op == "<":
        return a < b
    elif op == "==":
        return a == b
    else:
        return "âŒ Invalid Operator"



print(evaluate("+", 10, 5))    # 15
print(evaluate(">", 10, 5))    # True
print(evaluate("==", 10, 10))  # True
print(evaluate("/", 10, 2))    # 5.0
````

````python
#âœ… 1. Diamond Pattern Logic

n=9

for i in range(1,n+1):
    if i%2!=0:
        print(' '*((n-i)//2)+'*'*i)

for j in range(n-1,0,-1):
    if j%2!=0:
        print(' '*((n-j)//2)+'*'*j)
````

````python
n = 20  # Must be odd for a symmetric diamond

for i in range(1, n + 1):
    # Use i for top half, n - i for bottom half
    line = i if i <= (n // 2 + 1) else n - i + 1
    
    if line % 2 != 0:
        print(' ' * ((n - line) // 2) + '*' * line)
````

````python
#pascal triangle

from math import comb  # comb(n, r) gives nCr directly in Python 3.8+

# Number of rows in the triangle
n = 6

# Loop through each row
for i in range(n):
    # Print spaces to center-align the triangle
    print(" " * (n - i), end="")

    # Loop through each element in the row
    for j in range(i + 1):
        # Print combination nCr = comb(i, j)
        print(comb(i, j), end=" ")

    # Move to next line after each row
    print()
````

````python
# Number of rows to print
n = 6

# Outer loop for each row
for i in range(n):
    # Print leading spaces to center-align the triangle
    print(" " * (n - i),end='')

    # Initialize the first value in each row
    val = 1

    # Inner loop to calculate and print values in the current row
    for j in range(i + 1):
        print(val, end=" ")
        
        # Compute next value in the row using the previous one
        val = val * (i - j) // (j + 1)

    # Move to the next line after each row
    print()
````
