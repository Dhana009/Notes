🗣️ **Answer**:

🗣️ **Answer**:

**Immutable** data types cannot be modified after creation — like `int`, `float`, `str`, and `tuple`.

This impacts how data is passed around in functions, especially when dealing with shared objects in automation scripts."

## 🧠 Concept Breakdown

---

### 📌 **1. Arithmetic Operators**

Used for numeric calculations.

```python
a = 10
b = 3

print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.33 (float division)
print(a // b) # 3 (integer division)
print(a % b)  # 1 (remainder)
print(a ** b) # 1000 (exponent)
```

▶️ **QA Use Case**: Calculating response time gaps, retries, durations, etc.

---

### 📌 **2. Comparison Operators**

Used to compare values (always return `True` or `False`)

```python
x = 10
y = 20

print(x == y)  # False
print(x != y)  # True
print(x > y)   # False
print(x < y)   # True
print(x >= y)  # False
print(x <= y)  # True
```

▶️ **QA Use Case**: Comparing actual vs expected results.

```python
assert response["status_code"] == 200
```

---

### 📌 **3. Logical Operators**

Used to combine multiple boolean conditions.

```python
a = True
b = False

print(a and b)  # False
print(a or b)   # True
print(not a)    # False
```

▶️ **QA Use Case**: Building decision logic.

```python
if status == "pass" and exec_time < 1.0:
    print("Test passed quickly")
```

---

### 📌 **4. Assignment Operators**

Used to assign or update values.

```python
x = 10
x += 5   # same as x = x + 5
x *= 2   # x = x * 2
```

▶️ **QA Use Case**: Test counters, retry counts, status flags.

---

### 📌 **5. Membership Operators**

Used to check for existence in collections.

```python
tags = ["api", "login", "logout"]

print("api" in tags)     # True
print("auth" not in tags) # True
```

▶️ **QA Use Case**: Validating expected keys or values in response data.

---

### 📌 **6. Identity Operators**

Used to compare memory references, not values.

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)  # True (same object)
print(a is c)  # False (same content, different object)
print(a == c)  # True (value equality)
```

▶️ **QA Use Case**: Useful in debugging reference issues in test data handling.

---

## 🔧 Expressions

An **expression** is any combination of values, variables, and operators that evaluates to a value.

```python
result = (10 + 5) * 2 > 20 and True
# ➝ (15 * 2 = 30) > 20 and True ➝ True and True ➝ True
```

🧪 Expressions are at the **heart of assertions**, conditions, and branching logic in automation scripts.

🔶 What is Expression Precedence?
In Python (and most languages), when an expression has multiple operators, precedence determines which operator is evaluated first.

Think of it like math BODMAS:

Brackets → Orders (powers) → Division → Multiplication → Addition → Subtraction

Python also follows similar logic with more operators.

### 📌 **2. if…else – Dual Branching**

```python
if status_code == 200:
    print("Test Passed")
else:
    print("Test Failed")
```

✅ Use in test case pass/fail logic and setting result status.

---

### 📌 **3. if…elif…else – Multi-condition Checks**

```python
time = 1.4

if time < 1.0:
    print("Fast")
elif time < 2.0:
    print("Moderate")
else:
    print("Slow")
```

✅ **Use case**: Categorizing response times or classifying test results by criteria.

---

### 📌 **4. Nested Conditionals**

```python
status = "pass"
time = 0.9

if status == "pass":
    if time < 1.0:
        print("Passed quickly")
    else:
        print("Passed but slow")
else:
    print("Test Failed")
```

✅ Often used in automation when checking **multiple fields** from an API response or UI validation.

---

### 📌 **5. Ternary Operator (One-Liner If-Else)**

```python
result = "PASS" if status_code == 200 else "FAIL"
print(result)
```

✅ Use this in concise **report generation** or **summary logs**.

---

### 📌 **6. Logical Operators Inside Conditionals**

```python
if status == "pass" and time < 1.0:
    print("Excellent Test Run")
```

✅ Combine multiple validation conditions.

### 🔹 **Q3: What is the difference between `if`, `elif`, and `else`?**

🗣️ **Answer**:

* `if` checks the initial condition.
* `elif` allows us to check multiple other conditions if the first is false.
* `else` catches everything else when all previous conditions fail.

```python
a = {"TC001", "TC002"}
```

```python
print(a & b)  # {'TC002'}
```

#### ➤ **Difference (In A not in B)**

```python
print(a - b)  # {'TC001'}
```

#### ➤ **Symmetric Difference (In A or B, not both)**

```python
print(a ^ b)  # {'TC001', 'TC003'}
```

✅ Best for comparing test results: expected vs actual.

```python
if "TC001" in a:
```

# ✅ **Topic 2: Variable Scope in Python**

## 🎯 Why Scope Matters for QA

| QA Use Case           | Scope Impact                    |
| --------------------- | ------------------------------- |
| Test utilities        | Avoid polluting global state    |
| Environment configs   | Use global config wisely        |
| Fixture state         | Avoid variable leaks            |
| Debugging flaky tests | Scope errors cause side effects |

---

## 🔹 What is Scope?

**Scope** refers to where a variable is accessible.

### There are 4 levels of scope in Python (LEGB Rule):

| Scope Type        | Level                                | Example             |
| ----------------- | ------------------------------------ | ------------------- |
| **Local (L)**     | Inside a function                    | Local variables     |
| **Enclosing (E)** | Outer function (in nested functions) | Closures            |
| **Global (G)**    | Top-level script/module              | Global config/flags |
| **Built-in (B)**  | Python built-ins                     | `len()`, `print()`  |

---

## 🔸 1. **Local Scope** – Inside a Function

```python
def run_test():
    status = "pass"
    print(status)

run_test()
print(status)  # ❌ Error: status is not defined
```

✅ Use local variables to **isolate state**, especially inside test steps.

---

## 🔸 2. **Global Scope** – Outside All Functions

```python
env = "staging"  # Global variable

def test_env():
    print("Running in:", env)

test_env()
```

✅ Use for **environment setup**, global test settings.

---

## 🔸 3. **Modifying Global Variables – Using `global`**

```python
status = "fail"

def update_status():
    global status
    status = "pass"

update_status()
print(status)  # pass
```

✅ **Use with caution** — global variables can make test suites **unpredictable** if not managed carefully.

---

## 🔸 4. **Enclosing Scope – With `nonlocal`**

### ✅ Useful in **nested functions** (e.g., retry logic, logging wrappers)

```python
def outer():
    result = "fail"

    def inner():
        nonlocal result
        result = "pass"

    inner()
    print(result)

outer()  # pass
```

✅ Use in **decorators**, **test wrappers**, **retries**, etc.

---

## 🔸 5. Variable Shadowing – When Local Hides Global

```python
name = "global"

def greet():
    name = "local"
    print(name)

greet()      # local
print(name)  # global
```

✅ Avoid naming test variables the same as global ones — **can cause bugs**.

---

## 🔸 6. LEGB Scope Resolution Summary

| Level         | Example Code             | Can Modify?      |
| ------------- | ------------------------ | ---------------- |
| **Local**     | Inside current function  | ✅ Yes            |
| **Enclosing** | In outer function        | ✅ via `nonlocal` |
| **Global**    | Top of script/module     | ✅ via `global`   |
| **Built-in**  | Python default functions | ❌ Never touch    |

## 🔹 What is Comprehension in Python?

**Comprehension** is a concise way to create **lists**, **dictionaries**, or **sets** in one readable line using a loop.

````python
#Q3 – Secure Login Validator
#✅ Task:
#Write a function that takes two inputs: username and password, and checks:
#
#If username is not empty
#
#If password is at least 6 characters
#
#If both are valid, return "Login successful"
#
#Else, return specific error messages:
#
#If username is empty → "Username cannot be empty"
#
#If password is too short → "Password too short"
#
#If both are invalid → "Invalid login credentials"

def validate_login(username, password):
    invalids = ['',' ']
    return "Invalid login credentials" if username in invalids and len(password)<6 else \
            "Username cannot be empty" if username in invalids else \
            "Password too short" if len(password)<6 else "Login successful"

validate_login("dhan", "secret123")     #→ "Login successful"
validate_login("", "secret123")         #→ "Username cannot be empty"
validate_login("dhan", "123")           #→ "Password too short"
validate_login("", "")                  #→ "Invalid login credentials"
````
